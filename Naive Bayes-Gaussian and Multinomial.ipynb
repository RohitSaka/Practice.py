<h1 style="color:blue;">NAIVE BAYES CLASSIFIER ALGORITHM (Gaussian,Multinomial) using Wine Data Set</h1> 

```python
#importing Wine Dataset using SKLEARN
from sklearn import datasets
wine = datasets.load_wine()
```

```python
#Using Feature_names(variables) as the column
import pandas as pd
df = pd.DataFrame(wine.data,columns=wine.feature_names)
df.head()
```

```python
#wine.target will be the target Variable
df['target'] = wine.target
df[20:70]
```

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
train_X ,test_X , train_y , test_y = train_test_split(wine.data,wine.target,test_size=0.2,random_state=100)
```

```python
#Importing naives_bayes from sklearn Library (GaussianNB,MultinomialNB)

from sklearn.naive_bayes import GaussianNB,MultinomialNB
model = GaussianNB()
model.fit(train_X,train_y)
```

```python
#Complete Accuracy
model.score(test_X,test_y)
```

```python
#Using Multinomial Naive Bayes to check the accuracy score and it turned out be 83% accurate 
#with the change of test_size the model score changes
model = MultinomialNB()
model.fit(train_X,train_y)
```

```python
model.predict(test_X)
```

```python
model.score(test_X,test_y)
```
